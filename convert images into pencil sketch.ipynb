# **IMPORTING LIBRARIES**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
%matplotlib inline
print("Necessary packages included successfully!")
```

# **LOADING THE IMAGE**

```python
image=cv2.imread("/content/tom.png")
image = cv2.resize(image,(500,300))
image
```

```python
img_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

# **CONVERTING INTO GREY SCALE**

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray_image
```

```python
plt.imshow(gray_image)
```

# **INVERTING THE IMAGE**

```python
inverted_image = cv2.bitwise_not(gray_image)
inverted_image
```

```python
plt.imshow(inverted_image)
```

# **SMOOTHING THE IMAGE**

```python
img_smoothing = cv2.GaussianBlur(inverted_image, (21, 21),sigmaX=0, sigmaY=0)
img_smoothing
```

```python
def dodgeV2(x, y):
    return cv2.divide(x, 255 - y, scale=256.0)
```

# **INVERTING THE IMAGE TO GET THE PENCIL SKETCH**

```python
pencil_sketch = dodgeV2(gray_image, img_smoothing)
plt.imshow(pencil_sketch)
```

# **COMPARING THE ORIGINAL IMAGE WITH THE PENCIL SKETCH**

```python
display(Image.fromarray(img_rgb))
display(Image.fromarray(pencil_sketch))
```
